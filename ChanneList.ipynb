```python
from googleapiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow

# Set API key or OAuth credentials
API_KEY = "AIzaSyDI9UcnptDL8DpCehGL1NZJa_3j8jK9Wlw"  # If using API Key
CLIENT_SECRETS_FILE = "client_secret.json"  # If using OAuth

# Define the API service name and version
API_SERVICE_NAME = "youtube"
API_VERSION = "v3"

# Function to authenticate and build the API service
def build_service():
    flow = InstalledAppFlow.from_client_secrets_file(CLIENT_SECRETS_FILE, scopes=["https://www.googleapis.com/auth/youtube.readonly"])
    credentials = flow.run_console()
    return build(API_SERVICE_NAME, API_VERSION, credentials=credentials)

# Function to get the list of channels
def get_channel_list(api_service):
    channels = []
    next_page_token = None

    while True:
        request = api_service.channels().list(
            part="snippet",
            mine=True,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()

        channels.extend(response.get("items", []))
        next_page_token = response.get("nextPageToken")

        if not next_page_token:
            break

    return channels

def main():
    # Authenticate and build the service
    api_service = build_service()

    # Get the list of channels
    channels = get_channel_list(api_service)

    # Print channel titles
    for channel in channels:
        print(channel["snippet"]["title"])

if __name__ == "__main__":
    main()
```
